```python
penguins['bill_length_mm'] = penguins['bill_length_mm'].fillna(penguins['bill_length_mm'].mean())
penguins['bill_depth_mm'] = penguins['bill_depth_mm'].fillna(penguins['bill_depth_mm'].mean())
penguins['flipper_length_mm'] = penguins['flipper_length_mm'].fillna(penguins['flipper_length_mm'].mean())
penguins['body_mass_g'] = penguins['body_mass_g'].fillna(penguins['body_mass_g'].mean())
penguins['sex'] = penguins['sex'].fillna(penguins['sex'].mode()[0])
```

<h1>1. Which female and male species had the longest average flippers recorded on Biscoe Island in 2009?</h1>

```python
flipperLengthMale = penguins[(penguins['island'] == 'Biscoe') & (penguins['sex'] == 'male') & (penguins['year'] == 2009)]
flipperLengthMale.value_counts().sum()
```

```python
flipperLengthFemale = penguins[(penguins['island'] == 'Biscoe') & (penguins['sex'] == 'female') & (penguins['year'] == 2009)]
flipperLengthFemale.value_counts().sum()
```

```python
np.round(flipperLengthMale[flipperLengthMale['species'] == 'Gentoo']['flipper_length_mm'].mean())
```

```python
np.round(flipperLengthMale[flipperLengthMale['species'] == 'Adelie']['flipper_length_mm'].mean())
```

```python
np.round(flipperLengthMale[flipperLengthMale['species'] == 'Chinstrap']['flipper_length_mm'].mean())
```

<h2>#1 INSIGHT</h2>

<h3>Male Gentoo penguins have an average flipper length of 218mm, making them the longest flippers recorded on Biscoe Island in 2009, while no female penguins were analyzed.</h3>

<h1>7. In 2008 records, which penguin species had a bill length of less than 50 mm, a flipper length greater than 200 mm, and a bill depth of less than 18 mm?</h1>

```python
records = penguins[(penguins['year'] == 2008) & (penguins['bill_length_mm'] <= 50) & (penguins['flipper_length_mm'] >= 200) & (penguins['bill_depth_mm'] <= 18)]
records[records['species'] == 'Adelie'].value_counts().sum()
```

```python
records[records['species'] == 'Gentoo'].value_counts().sum()
```

```python
records[records['species'] == 'Chinstrap'].value_counts().sum()
```

<h2>#7 INSIGHT</h2>

<h3>1 Adelie and 39 Gentoo penguins had a bill length of less than 50 mm, a flipper length greater than 200 mm, and a bill depth of less than 18 mm, while no Chinstrap penguins met these criteria.</h3>
